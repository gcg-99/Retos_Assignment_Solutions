```ruby
def find_orthologues(tair_homologues_file,spom_homologues_file)
  File.open("orthologues_list.txt", "w") { |f| f.write("Pairs of possible orthologues between A. thaliana and S. pombe:\n") }  
  
  tair_homologues = []
  spom_homologues = []
  orthologues_list = []
  
  file = File.read(tair_homologues_file).split("\n")
  file.each do |pair|
    field = pair.split("\t")
    tair_homologues << [field[0],field[1]]
  end
  
  file = File.read(spom_homologues_file).split("\n")
  file.each do |pair|
    field = pair.split("\t")
    spom_homologues << [field[0],field[1]]
  end
  
  
  tair_homologues.each do |tair_hom|
    spom_homologues.each do |spom_hom|
      if tair_hom[0] == spom_hom[1] and tair_hom[1] == spom_hom[0]
        puts "Orthologue found!"
        puts "SPOM: #{spom_hom}"
        puts "TAIR: #{tair_hom}"
        orthologues_list << spom_hom
      end
    end
  end
  
  orthologues_list = orthologues_list.uniq #removing possible duplicated pairs of orthologues (just in case, but it should not be any)
  orthologues_list.each do |line|
    File.open("orthologues_list.txt", "a") { |f| f.write("\t#{line[0]} & #{line[1]}\n") }
  end
end
```

```ruby
find_orthologues("./best_tair.txt","./best_spom.txt")
```
